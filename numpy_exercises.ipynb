```python
import numpy as np
```

you need to track sales of 3 products(headphones,watches,speakers) over 4 days(monday,tuesday,wednesday,thursday)
row 0 headphones
row 1 smartphones
row 2 speakers

```python
cost_price = np.array([1500,3000,5000,0])

daily_sales = np.array([[10,15,8,12],[5,7,4,6],[2,3,2,4]])

print(np.shape(cost_price)) #prints the shape of the cost_price array which is (4,) because it is a 1-dimensional array with 4 elements
print(np.shape(daily_sales)) #prints the shape of the daily_sales array which is (3, 4) because it is a 2-dimensional array with 3 rows and 4 columns
print(np.ndim(cost_price))
print(np.ndim(daily_sales))
```

manager decides to add 2 bonus units to  every product of products daily sale for a report. add 2 to entire daily sales array and then calculate total revenue. then multiply the cost_price array by the daily_sales matrix. Observe how numpy handles the multiplication of 1d and 2d array.

```python
daily_sales_bonus = daily_sales + 2

print("\nAfter Adding Bonus Units:\n", daily_sales_bonus)

cost_price = np.array([2000, 50000, 8000])  # per unit

print("\nCost Price:\n", cost_price)

revenue_matrix = daily_sales_bonus * cost_price[:, np.newaxis]

print("\nRevenue Matrix:\n", revenue_matrix)

total_revenue = np.sum(revenue_matrix)

print("\nTotal Revenue =", total_revenue)
```
