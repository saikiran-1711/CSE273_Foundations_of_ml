# Standard deviation


st deviations tells the distance of data from mean

ex: student commute avg is 30min

Student A has Stdev = 2 so he arrives at (28-32min)highly predictable.

Student B has Stdev = 15 so he arrives in between (15-45 min) unpredictable.


## The Bell curve


when the St dev is in range out +3 to -3 st deviations from mean it is under curve and expected. if not it is highly unusual

### Z-Score:
