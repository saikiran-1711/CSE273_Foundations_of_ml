```python
import numpy as np

```

```python
marks = np.array([
    [78, 85, 90],   # Student 1
    [45, 60, 55],   # Student 2
    [88, 92, 95],   # Student 3
    [30, 40, 35],   # Student 4
    [65, 70, 75]    # Student 5
])

print(marks)

```

```python
print(np.shape(marks))  # Shape of the array
print(np.size(marks))   # Total number of elements
print(np.result_type(marks))  # Data type of the elements

#column and row numbering starts from 0
print(marks[1]) # Accessing the second row (Student 2's marks)
print(marks[0:,2])     # marks of all students in the third subject
print(marks[0, 1])#  marks of first student in second subject



```

```python
print(np.mean(marks, axis=0))  # Average marks for each subject
avg = np.mean(marks, axis=1)  # Average marks for each student
result = np.where(marks >= 80, 1, 0)
print(result)


#global adjustments
print(marks+5) # Adding 5 marks to each student's score
print(marks) 

```

```python
mapping = np.where(marks > 50,1,0)
print(mapping)

data_cleaning = np.where(marks > 40,1,0)
print(data_cleaning)
```

```python
Grade_allocation = np.where(avg>=85,'A',np.where(avg>=70,'B',np.where(avg>=50,'C','No grade')))

# The above code is using nested np.where statements to allocate grades based on the average marks of each student.
# It checks if the average marks are greater than or equal to 85, then assigns 'A'. If not, it checks if the average marks are greater than or equal to 70,

print(Grade_allocation)
```
