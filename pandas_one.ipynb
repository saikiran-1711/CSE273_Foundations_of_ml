Series from Dictionary

```python
#series of python to pandaas
data = {'Maths': 90, 'Science': 85, 'English': 85}
print(data)
marks = pd.Series(data)
print(marks)
#why no index here? because it is a dict, so it will take the keys as index and values as values
```

```python
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Marks": [85, 90, 95]
}
print(data) #python dictionary
print(type(data))
#convert to pandas dataframe
marks = pd.DataFrame(data) #pandas automatically adds a default index starting from 0 by default. this is called RangeIndex

print(marks)
```

```python
#we can also add custom index in pandas dataframe
marks = pd.DataFrame(data, 
        index=['a', 'b', 'c']
        )
print(marks)
```

```python
marks.set_index("Name", inplace=True)
#if we are setting inplace as true it means we are changing the original dataframe, 
# if we are not setting inplace as true it means we are creating a new dataframe with the new index and the original dataframe will remain unchanged
```

```python
df.isnull()
#prints true is null and false if not null
```
