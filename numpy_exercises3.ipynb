```python
#task1: create an array of random integers between 120 and 650 with a shape of (4,7)
electricity = np.array([np.random.randint(120,650,(4,7))]) #
print(electricity)

print(electricity.shape)
print(electricity.ndim)
print(electricity.size)
print(electricity.dtype)  
```

```python
#calculate the average electricity consumption of each hostel
avg_electricity_per_hostel = np.mean(electricity[:], axis=1) #perform mean along the rows

print(avg_electricity_per_hostel) #print avg electricity of each hostel

print(max(avg_electricity_per_hostel))
max_electricity = np.where(avg_electricity_per_hostel == max(avg_electricity_per_hostel),1,0)
print(max_electricity)

electricity_consumption_everyday = np.mean(electricity[0], axis=0)
print(electricity_consumption_everyday)



```

```python
#if electricity_consumption_everyday[0] > 550 then surcharge of 20% is applied
surcharge = np.where(electricity_consumption_everyday > 550, electricity_consumption_everyday * 0.2, 0)
print(surcharge)

arr = electricity.copy()
print(arr)

penalty_array =np.where(arr > 550, arr * 1.2,arr)
print(penalty_array)

check = np.where(penalty_array == arr, 0,1)
print(check)
```

```python
daily_avg = np.where(electricity > avg_electricity_per_hostel,"True","False")
print(daily_avg)
```

```python
#reshape the electricty data into single row representing weekly usage(shape 1*28)
weekly_usage = electricity.reshape(1,28)
print(weekly_usage)

#reshape the electricty data into 2 rows and 14 columns representing. no data is lost or dupliacted
reshaped_electricity = electricity.reshape(2,14)
print(reshaped_electricity)
```
